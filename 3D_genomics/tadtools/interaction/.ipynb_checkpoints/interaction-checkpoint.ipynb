```python
import pandas as pd
import statistics
```

```python
interaction = pd.read_csv('KK_Bonev_Nora_pixels.mat',sep='\t',header=None,names=['chrom1','start1','end1','chrom2','start2','end2','notsure','v1','v2','v3'])
```

```python
interaction.head()
```

```python
interaction['avg'] = interaction[['v1','v2','v3']].mean(axis=1)
```

```python
interaction.head()
```

```python
tads = pd.read_csv('merge_TADs.bed',sep='\t')
```

```python
tads.head()
```

```python
chroms = ['chr'+str(i) for i in range(1,20)]+['chrX']
#tad_fpkm_chrs = []
tads['interaction'] = ''
for chrom in chroms:
    tads_chr = tads.loc[tads.chrom==chrom]
    interaction_chr = interaction.loc[(interaction.chrom1==chrom) & (interaction.chrom2==chrom)]
    #fpkm_tad_means = []
    for idx, (c,s,e,i) in tads_chr.iterrows():
        tad_inter = interaction_chr.loc[(interaction_chr.start1==s) & (interaction_chr.start2==e),'avg']
        num = len(tad_inter.to_numpy())
        if num==1:
            tads.iloc[idx,3] = tad_inter.to_numpy()
```

```python
tads.to_csv('merge_TADs_interaction.tsv',sep='\t',index=False,quoting=3)
```
